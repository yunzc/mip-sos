### Iterate through obstacle configurations (at most 90 based on our discretization)
Now need to construct a library of trajectories. And store for online use. 

```python
# investigate time out 
OA = ObstAvoid()
obst_idx = np.zeros(3)
obst_idx[0:2] = 2

x = Array('d', range(31))
y = Array('d', range(31))
ux = Array('d', range(30))
uy = Array('d', range(30))

p = Process(target=OA.compute_trajectory, name="Compute_traj", args=(obst_idx, x, y, ux, uy))
p.start()
# Wait 5 seconds
time.sleep(5)
timeout = False
if p.is_alive():
    print "timeout solver"
    timeout = True
    p.terminate()
    p.join()

traj = np.zeros([31,2])
traj[:,0] = x[:]
traj[:,1] = y[:]
OA.plot_trajectory(traj, obst_idx)
print('timeout: ', timeout)
```

```python
# ang_discret=5, rng_discret=5 at most 3125 trajectories 
# saved datas 
l = 5**5
obst_array = np.zeros([l,5])
x_array = np.zeros([l,31])
y_array = np.zeros([l,31])
ux_array = np.zeros([l,30])
uy_array = np.zeros([l,30])
OA = ObstAvoid()
i = 0
for a in range(5):
    for b in range(5):
        for c in range(5):
            for d in range(5):
                for e in range(5):
                    obst_idx = [a,b,c,d,e]
                    x = Array('d', range(31))
                    y = Array('d', range(31))
                    ux = Array('d', range(30))
                    uy = Array('d', range(30))

                    p = Process(target=OA.compute_trajectory, name="Compute_traj", args=(obst_idx, x, y, ux, uy))
                    p.start()
                    # Wait 5 seconds
                    time.sleep(5)
                    timeout = False
                    if p.is_alive():
                        print "timeout solver"
                        timeout = True
                        p.terminate()
                        p.join()
                    obst_array[i,:] = obst_idx
                    if timeout:
                        x_array[i,:] = np.zeros([31])
                        y_array[i,:] = np.zeros([31])
                        ux_array[i,:] = np.zeros([30])
                        uy_array[i,:] = np.zeros([30])
                    else:
                        x_array[i,:] = x[:]
                        y_array[i,:] = y[:]
                        ux_array[i,:] = ux[:]
                        uy_array[i,:] = uy[:]
                    i += 1
                    print("%s of 3125"%(str(i)))
                    
np.savetxt('x.out', x_array, delimiter=',')
np.savetxt('y.out', y_array, delimiter=',')
np.savetxt('ux.out', ux_array, delimiter=',')
np.savetxt('uy.out', uy_array, delimiter=',')
np.savetxt('obst.out', obst_array, delimiter=',')
```

### Control Design along Trajectories with SOS 
First step is to find a starting Lyapunov

```python
from pydrake.all import LinearQuadraticRegulator
A = np.array([[0,0,0],[0,0,2.5],[0,0,0]])
B = np.eye(3)

def create_lqr(A, B):
    Q = 10*np.eye(3)
    Q[2,2] = 1.5
    R = np.eye(3)
    K, S = LinearQuadraticRegulator(A, B, Q, R)
    return (K, S)

K, S = create_lqr(A, B)
print "LQR K: ", K
print "LQR S: ", S

# xdot = Ax + Bu

# V = x.T S x (matmul)
```

### Find region of attraction 

```python
# test LQR
xn = np.array([0.2,0.4,0]).reshape([3,1])
u = np.matmul(K, xn)
print(u)
```

```python
xf = np.array([0.125, 0.025, 0])
# Calculates the closed loop f(x) at xn
def plant(x, u):
    xd = np.cos(x[-1])*u[0] - np.sin(x[-1])*u[1]
    yd = np.sin(x[-1])*u[0] + np.cos(x[-1])*u[1]
    thetd = u[2]
    return np.array([xd,yd,thetd])

def calcF(xn):
    # K from lqr
    global K
    global xf
    xn = np.array(xn).reshape([-1,1])
    x_ = xn - xf
    un = np.matmul(K, xn)
    f = plant(un,xn)
    return f

# Calculates V(xn)
def calcV(xn):
    # S from LQR
    global S
    global xf
    x0 = xf[:].reshape([3,1])
    x = np.array(xn).reshape([3,1])
    x_ = x - x0
    Sx = np.matmul(S,x_)
    V = np.matmul(x_.T, Sx)[0,0]
    return V

# Calculates \dot{V}(xn).
def calcVdot(xn):
    # S from LQR
    global S
    global xf
    x0 = xf[:].reshape([3,1])
    x = np.array(xn).reshape([3,1])
    x_ = x - x0
    dvdx = 2*np.matmul(x_.T,S)
    f = calcF(xn)
    vdot = np.matmul(dvdx, f)
    return vdot[0]

start_time = time.time()

x0 = np.zeros([3,1])

# Sample f, V, and Vdot over
# a grid defined by these parameters.
# (Odd numbers are good because there'll be
# a bin at exactly the origin.
n_bins = 31
n_bins_theta = 17
# For theta and thetad, we only need to span
# a small region around the fixed point
width = 10
width = 10
# For \dot{theta_2}, though, the default
# parameters for our pendulum lead us to
# need to search larger absolute \dot{theta_2}
# values (because the inertial wheel is relatively
# light).
theta_width = 10

# Do the actual sampling....
x = np.linspace(x0[0]-width, xf[0]+theta_width, n_bins)
y = np.linspace(xf[1]-width, xf[1]+width, n_bins)
thet = np.linspace(xf[2]-theta_width, xf[2]+theta_width, n_bins_theta)
X, Y, Thet = np.meshgrid(x, y, thet, indexing="ij")

def calc_over_array(f):
    return np.array([[[f([dx, dy, dthet]) for dthet in thet] for dx in x] for dy in y])

V_samples = calc_over_array(calcV)
f_samples = calc_over_array(calcF)
Vdot_samples = calc_over_array(calcVdot)

elapsed = time.time() - start_time
print "Computed %d x %d x %d sampling in %f seconds" % (n_bins, n_bins, n_bins_theta, elapsed)

```

#### Find rho through samping 

```python
def estimate_rho(V, Vdot):
    '''
    Code submission for 3.6
    Fill in this function to use the samples of V and Vdot
    (Each array has dimension [n_bins, n_bins, n_bins_theta2d])
    to compute a maximal rho indicating the region of attraction
    of the fixed point at the upright.
    '''
    rho = np.inf
    for i in range(n_bins):
        for j in range(n_bins):
            for k in range(n_bins_theta2d):
                if Vdot[i,j,k] >= 0:
                    if V[i,j,k] < rho:
                        if Vdot[i,j,k] != 0 and V[i,j,k] != 0:
                            rho = V[i,j,k]
    return rho*0.9 # account for descretizations 

rho = estimate_rho(V_samples, Vdot_samples)
print "Region of attraction estimated at V(x) <= ", rho
```
